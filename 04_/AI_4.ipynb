```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import graphviz
f = open('res.txt')


x = []
y = []
i = 0

for s in f:
  s = s.replace('\n', '')
  x.append([float(x) for x in list(s)])
  y.append(i)
  i+=1
  
clf = DecisionTreeClassifier()
clf.fit(x, y)

import graphviz
from sklearn import tree
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("mytree")
```

```python
for s in x:
  p = np.array(s)
  print(p.reshape(-1, 5))
```

```python
f = open('source.dat')
d = []
for s in f:
  d.append([float(x) for x in s.split()])
d = np.array(d)

x = d[:, :-1]
y = d[:, -1]
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

```python
clf = DecisionTreeClassifier()
clf.fit(x, y)

dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("mytree")
```

```python
x2 = x.copy()
x2[:, 0] = x2[:, 0] + 0.2 * x2[:, 1] #    PCA
plt.scatter(x2[:,0], x2[:,1], c=y) #   
```

```python
clf = DecisionTreeClassifier()
clf.fit(x2, y)

dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("mytree")
```

```python
from sklearn.decomposition import PCA 
clf = PCA()
x2 = clf.fit_transform(d)
plt.scatter(x2[:,0], x2[:,], c=y)
```
