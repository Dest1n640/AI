```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def circle(x, y, r):
    if x**2 + y**2 <= r**2:
        return 1
    else:
        return 0

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
xx, yy = np.meshgrid(x, y)
colors = np.array([circle(x, y, 1) for x, y in zip(xx.ravel(), yy.ravel())])

plt.scatter(xx.ravel(), yy.ravel(), c=colors)
plt.show()
```

```python
def c(x, y):
  return x**2 + y**2 < 1

X = (np.random.rand(10000, 2) - 0.5) * 3

plt.scatter(X[:, 0], X[:, 1], c=c(X[:, 0], X[:, 1]), s=4)
plt.show()
```
