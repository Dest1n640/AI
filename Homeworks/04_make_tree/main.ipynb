```python
import numpy as np 
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
```

```python
f = open('test_add.dat')
d = []
for l in f:
  d.append([float(x) for x in l.split()])
d = np.array(d)

x = d[:, :-1]
y = d[:, -1]
x, y
```

```python
depth = np.arange(1, 30010, 1000)
min_leaf = [0.0] #Этот параметр всегда будет уменьшать score, поэтому оставляю 0.0
max_node = np.arange(2, 30010, 1000)


best_score = 0.0
best_param = None
best_clf = None

high_scores = []
high_scores_params = []

for i in depth:
    for j in min_leaf:
        for k in max_node:
            clf = DecisionTreeClassifier(max_depth=i, min_weight_fraction_leaf=j, max_leaf_nodes=k)
            clf.fit(x, y)
            score = clf.score(x, y)
            if score > 0.998:
              high_scores.append(score)
              high_scores_params.append((i, j, k))
            
            print(f"Пробую параметры: depth={i}, leaf={j}, nodes={k}. Результат: {score}")

            if best_score < score:
                best_score = score
                best_param = (i, j, k)
                best_clf = clf
                print(f"score изменился - {best_score}")
                
print("Оценки > 0.998:", high_scores)
print("Параметры для этих оценок:", high_scores_params)
```

```python
print(high_scores)
print(high_scores_params)
#По выводу данных мы можем сделать вывод 
#Что максимально score мы досгиаем благодаря nodes >= 30000 при глубине >=1000
#Поэтому лучшую модель мы будем строить с учетом этого
```

```python
# Доказательство того что leaf уменьшает общий score
leaf = np.arange(0.1, 0.51, 0.1)
for i in leaf:
  clf = DecisionTreeClassifier(max_depth = 1000, max_leaf_nodes = 30000, min_weight_fraction_leaf = i)
  clf.fit(x, y)
  score = clf.score(x, y)
  print(f"Пробую параметры: depth={1000}, leaf={i}, nodes={30000}. Результат: {score}")
  
```

### С учетом, того что лучший скор достигается путем изменение только одного параметра(Такой вывод был сделан на основе анализа вывода данных), можно предположить что можно достичь скора при изменении максимальной глубины

```python
depth = np.arange(1, 1000, 10)
best_depth = 0
best_score = 0
for i in depth:
  clf = DecisionTreeClassifier(max_depth=i)
  clf.fit(x, y)
  score = clf.score(x, y)
  if score == 1:
    best_score = score
    best_depth = i
    break
print(f"Score равный единице найден - {best_score}")
print(f"Score равен одну при max_depth = {best_depth}")
```

```python
max_leaf = np.arange(2, 30100, 1000)
best_score = 0
best_leaf = 0
for i in max_leaf:
  clf = DecisionTreeClassifier(max_leaf_nodes=i)
  clf.fit(x, y)
  score = clf.score(x, y)
  if score >= 0.999:
    best_score = score
    best_leaf = i
    print(f"Найден score >= 0.999 - {best_score}")
    print(f"Score достигает значения {best_score} при max_leaf_nodes = {best_leaf}")
```

### На основе прошлой гипотезы предпологаю так же что только при изменении min_weight_fraction_leaf score тоже будет близок к единице

```python
leaf = np.arange(0.1, 0.5, 0.1)
best_score = 0
best_weight = 0
for i in leaf:
  clf = DecisionTreeClassifier(min_weight_fraction_leaf=i)
  clf.fit(x, y)
  score = clf.score(x, y)
  if score == 1:
    best_score = score
    best_weight = i
if best_score <= 0.998:
  print("Результат не найден")
else:
    print(f"Найден score >= 0.999 - {best_score}")
    print(f"Score достигает значения {best_score} при min_weight_fraction_leaf = {best_weight}")
# С изменением минимального единицы никогда не достигнет 1
# Моя гипотеза неверна
    
```
