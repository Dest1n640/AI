```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import graphviz
```

```python
# SVD -   

f = open('source.dat')
d = []
for s in f:
  d.append([float(x) for x in s.split()])
d = np.array(d)

x = d[:, :-1]
y = d[:, -1]
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
x = d[:, :2]
y = d[:,2]
clf.fit(x, y)
yp = clf.predict(x)
fig, axs = plt.subplots(1, 2)
axs[0].scatter(x[:,0], x[:, 1], c =y)
axs[0].set_title('Labels')
axs[1].scatter(x[:,0], x[:, 1], c =yp)
axs[1].set_title('Pred')
plt.show()
```

```python
print(f"N: {clf.coef_}")
print(f"B: {clf.intercept_}")
```

```python
def sigm(x):
  return 1 / (1 + np.e ** (-x))

plt.plot(x, sigm(x))
plt.show()
```

```python
def gip_tg(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
def gip_tg_norm(x):
  return np.tanh(x)
plt.plot(x, gip_tg_norm(x))
plt.show()
```

```python
def relu(x):
  c = x.copy()
  c[c<0] = 0
  return c
# return x * ((np.sign(x) + 1) / 2)
plt.plot(x, relu(x))
plt.show()
```
